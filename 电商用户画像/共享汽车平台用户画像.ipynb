```python
#导入工具包
import numpy as np
import pandas as pd
from datetime import datetime
import time
```

```python
#导入数据
data1=pd.read_excel('用户首单表.xlsx')
data2=pd.read_excel('用户整体分层.xlsx')
data=pd.DataFrame(pd.merge(data2,data1,on='用户编号',how='left'))
```

```python
N=data.iloc[:,0].size
```

```python
data.info()
```

```python
data['身份证'].apply(lambda x:1 if len(str(x))<18 else 0).sum()
```

```python
#填充缺失值
data['年']=data['年'].fillna(0).astype('int')
T=list()
for i in range(N):
    if data['首单时间'][i]=='1970-01-01 08:00:00':
        T.append('2019-01-01 08:00:00')
    elif data['首单时间'][i]=='NaN' or data['首单时间'][i]=='':
        T.append('0000-00-00 00:00:00')
    else:
        T.append(data['首单时间'][i])
data['首单时间']=T

S=list()
for i in range(N):
    if len(str(data['身份证'][i]))<18:
        S.append('000000000000000000')
    else:
        S.append(data['身份证'][i])
data['身份证']=S
data['身份证长度']=[len(i) for i in data['身份证'].tolist()]
```

```python
data[['FF','MM']].describe()
```

```python
#数据处理
data['出生年']=data['身份证'].str[6:10]
data['出生日期']=data['身份证'].str[6:14]
data['sex']=data['身份证'].str[-2]
data['当前年']=int(datetime.now().strftime('%Y'))
data['月数']=(data['当前年']-data['年'])*12
data['年龄']=data['当前年']-data['出生年'].astype('int')
data['单均价']=round(data['MM']/data['FF'])
data['条件']=(data['优惠券使用次数']/data['FF']).astype('float')
dd='2019-05-01 14:40:40'
data['最近活跃']=(pd.to_datetime(dd)-pd.to_datetime(data.RR)).dt.days
#当数据存在极值时，应用中位数表示数据的中间值
R_median=np.median(data.最近活跃)
F_median=np.median(data.FF)
M_median=np.median(data.MM)
data['R']=np.where(data.最近活跃<R_median,'高','低')
data['F']=np.where(data.FF>F_median,'高','低')
data['M']=np.where(data.MM>M_median,'高','低')
data['RFM']=data.R+data.F+data.M
```

```python
data.年龄.max()
```

```python
#用户标签
data['用户ID']=data['用户编号']
data['性别']=['女' if float(i)%2 ==0 else '男' for i in data.sex.tolist()]
data['年龄分段']=pd.cut(data['年龄'],bins=[0,19,24,29,34,39,44,49,54,59,2020],
                   labels=['<20','20-25','25-30','30-35','35-40','40-45','45-50','50-55','55-60','60+'])
def func(x):
    if x>=1:
        return '全用'
    elif x>0 and x<1:
        return '半用'
    else:
        return '不用'
data['使用优惠券情况']=data['条件'].apply(func)

def func2(x):
    if x<=0.5:
        return '非工作日活跃'
    elif x>0.5 :
        return '工作日活跃'
    else:
        return '全周活跃'
data['活跃情况']=(data['周中次数']/data['FF']).apply(func2)

def func3(x):
    if x<=0.5:
        return '下午活跃'
    elif x>0.5:
        return '上午活跃'
    else:
        return '全天活跃'
data['上下午活跃情况']=(data['上午通勤次数']/data['FF']).apply(func3)

data['活跃集中情况']=(data['早高峰']/data['FF']).apply(lambda x: '早峰活跃' if x<=0.5 else '晚峰活跃')

D=[]
for i in range(N):
    if data.FF[i]==1 and data.最近活跃[i]<=15:
        D.append('引入期')
    elif data.FF[i]>1 and data.FF[i]<=3 and data.最近活跃[i]<=70:
        D.append('成长期')
    elif data.FF[i]>=4 and data.最近活跃[i]<=70:
        D.append('成熟期')
    elif data.最近活跃[i]>=70 and data.最近活跃[i]<=90:
        D.append('休眠期')
    elif data.最近活跃[i]>=90 or (data.FF[i]==1 and data.最近活跃[i]>15):
        D.append('流失期')
    else:
        D.append('else')
data['生命周期']=D

data['月均单量情况']=np.where(data['FF']/data['月数']>=6,'月均单数大于6','月均单数小于6')
data['单均价情况']=np.where(data['单均价']>=40,'单均价大于40','单均价小于40')
data['长短途情况']=np.where(data['总里程']/data['FF']>=30,'长途居多','短途居多')
data['优惠券金额占比']=round(data['优惠金额总数']/(data['优惠金额总数']+data['MM']),4)
data['时租占比']=np.where(data['时租次数']/data['FF']>0.5,'时租居多','日租居多')
data['取车偏好']=np.where(data['自取次数']/data['FF']>0.5,'自取居多','送车居多')
data['还车偏好']=np.where(data['非网点还车次数']/data['FF']>0.5,'网点外还车居多','网点还车居多')
RFM={
    '高高高':'重要价值用户',
    '高低高':'重要保持用户',
    '低高高':'重要发展用户',
    '低低高':'重要保持用户',
    '高高低':'一般价值用户',
    '高低低':'一般发展用户',
    '低高低':'一般维持用户',
    '低低低':'一般挽留用户'
}
data['RFMlevel']=[RFM[i] for i in data.RFM.tolist()]
    
```
