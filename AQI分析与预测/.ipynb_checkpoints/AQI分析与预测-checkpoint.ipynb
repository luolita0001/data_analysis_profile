### 5.全国城市空气质量普遍处于何种水平

```python
#城市空气质量等级划分
# 0-50 一级 优
# 51-100 二级 良
# 101-150 三级 轻度污染
# 151-200 四级 中度污染
# 201-300 五级 重度污染
# >300 六级 严重污染
def AQI_func(x):
    if 0<=x and x<=50:
        return '一级'
    elif 51<=x and x<=100:
        return '二级'
    elif 101<=x and x<=150:
        return '三级'
    elif 151<=x and x<=200:
        return '四级'
    elif 201<=x and x<=300:
        return '五级'
    else:
        return '六级'
level=data['AQI'].apply(AQI_func)
print(level.value_counts())
```

```python
sns.countplot(x=level,order=['一级','二级','三级','四级','五级'])
plt.show()
```

- 可见城市的空气质量主要以一级(优)和二级(良)为主，三级（轻度污染）占一部分，更高污染的城市占少数

#### 假设检验
- 采用两独立样本t检验进行验证
- 两独立样本t检验需要满足的前提条件是:
    - 两样本之间是独立的
    - 两样本服从正态分布

#### 正态分布检验
- 绘制数据的分布图
- 绘制pp图和qq图
- 使用假设检验

```python
#绘制数据的分布图
coastal=data[data['Coastal']=='是']['AQI']
inland=data[data['Coastal']=='否']['AQI']
fig,ax=plt.subplots(1,2)
fig.set_size_inches(15,5)
sns.distplot(coastal,ax=ax[0])
sns.distplot(coastal,ax=ax[1])
plt.show()
#
```

- 两边都呈现明显右偏

```python
from scipy import stats
#pp图和qq图
#pp图：通过累积概率来检测
#qq图：通过分布百分比对应的数值来检测
import statsmodels.api as sm
def plot_pp_qq(data):
    '''
    data:要绘制的数值
    '''
    fig,ax=plt.subplots(1,2)
    fig.set_size_inches(15,5)
    #进行标准化处理
    scale_data=(data-data.mean())/data.std()
    p=sm.ProbPlot(data=scale_data,dist=stats.norm)#dist分布
    p.ppplot(line='45',ax=ax[0])
    ax[0].set_title('PP图')
    p.qqplot(line='45',ax=ax[1])
    ax[1].set_title('QQ图')
    plt.show()
```

```python
plot_pp_qq(coastal)
```

```python
plot_pp_qq(inland)
```

- 可以看出两样本均偏离正态分布

```python
#假设检验
print(stats.normaltest(coastal))
print(stats.normaltest(inland))
```

- p值均小于0.05,应拒绝原假设，认为两样本均不服从正态分布

#### 非正态总体的检验
- 将分布转换为正态分布
    - 使用box_cox变换转换为正态分布
- 使用非参数检验
    - 使用曼-惠特尼检验
    - 威尔科克森秩和检验

##### 将分布转换为正态分布
- Box-Cox变换

```python
#输入:需要转换的数据
#输出:1.转换之后的结果，2.最合适的lambda值，即能够使转换结果最符合正态分布的lambda值
be_coastal,_=stats.boxcox(coastal)
be_inland,_=stats.boxcox(inland)
```

- 正态分布检验

```python
fig,ax=plt.subplots(1,2)
fig.set_size_inches(15,5)
sns.distplot(be_coastal,ax=ax[0])
sns.distplot(be_inland,ax=ax[1])
plt.show()
```

```python
plot_pp_qq(be_coastal)
```

```python
plot_pp_qq(be_inland)
```

```python
print(stats.normaltest(be_coastal))
print(stats.normaltest(be_inland))
#p值均大于0.05，即无法拒绝原假设，即认为转换后服从正态分布
```

- 两独立样本t检验

```python
#查看两样本的方差是否齐性
stats.levene(be_coastal,be_inland)
```

```python
#连起来，此时的p值是双侧检验得到的p值
r=stats.ttest_ind(be_coastal,be_inland,equal_var=stats.levene(be_coastal,be_inland).pvalue>0.05)
print(r)
```

```python
#原假设：临海城市的空气质量好于内陆城市，<
#计算t统计量时，是be_coastal-be_inland,即原假设成立，t统计量越小，p值应该越大，即使用sf
stats.t.sf(r.statistic,df=len(be_coastal)+len(be_inland)-2)
```

```python
#等同于
stats.t.cdf(abs(r.statistic),df=len(be_coastal)+len(be_inland)-2)
```

### 6.怎样预测一个城市的空气质量
- 数据转换
- 建立基模型
- 处理异常值
- 数据标准化
- 特征选择

- 残差图分析

#### 数据转换
- 在建模时需要将类别变量转换为离散变量
- 二分类变量可以直接转换为0，1
- 多分类变量应使用热编码one-hot的方式进行转换

```python
data['Coastal']=data['Coastal'].map({'是':1,'否':0})
data['Coastal'].value_counts()
```

#### 基模型
- 后续的操作可以在基模型中进行

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#city字段对预测结果无用处，故删除
X=data.drop(['AQI','City'],axis=1)
y=data['AQI']
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)

lr=LinearRegression()
lr.fit(x_train,y_train)
print('训练数据的得分',lr.score(x_train,y_train))
print('测试数据的得分',lr.score(x_test,y_test))
```

```python
y_hat=lr.predict(x_test)
plt.figure(figsize=(15,5))
plt.plot(y_test.values,'-r',label='真实值',marker='o')
plt.plot(y_hat,'-g',label='预测值',marker='*')
plt.legend(loc='upper left')
plt.title('线性回归预测结果',fontsize=20)
plt.show()
```

#### 处理异常值
- 数据存在异常值可能会影响模型的效果，在建模时可以利用箱线图去判断异常值，在使用临界值替换异常值

```python
numerical_features.remove('AQI')
```

```python
numerical_features
```

```python
np.quantile(x_train['Precipitation'],[0.25,0.75])
```

```python
#Coastal是类别变量不存在异常值，因此仅需要对数值型变量进行异常值处理
for col in numerical_features:
    if pd.api.types.is_numeric_dtype(x_train[col]):
        quartile=np.quantile(x_train[col],[0.25,0.75])
        IQR=quartile[1]-quartile[0]
        lower=quartile[0]-1.5*IQR
        upper=quartile[1]+1.5*IQR
        x_train[col][x_train[col]<lower]=lower
        x_train[col][x_train[col]>upper]=upper
        x_test[col][x_test[col]<lower]=lower
        x_test[col][x_test[col]>upper]=upper
```

```python
#效果对比
lr.fit(x_train,y_train)
print('训练数据的得分',lr.score(x_train,y_train))
print('测试数据的得分',lr.score(x_test,y_test))
#得分略有所提高
```

#### 数据标准化
- 数据标准化不会影响评估效果，但考虑权重代表的实际意义，可以对数据进行标准化处理

```python
from sklearn.preprocessing import StandardScaler
s=StandardScaler()
x_train_scale=s.fit_transform(x_train)
x_test_scale=s.fit_transform(x_test)
```

#### 特征选择
- RFECV特征选择
- RFE阶段：
    - 1.初始的特征集为所有可用的特征
    - 2.使用当前特征集进行建模，然后计算每个特征的重要性
    - 3.删除最不重要的一个或多个特征，更新特征集
    - 4.跳转到步骤2，直到完成所有特征的重要性评级
- CV阶段
    - 1.根据RFE阶段确定的特征重要性，依次选择不同数量的特征
    - 2.对选定的特征集进行交叉验证
    - 3.确定平均分最高的特征数量，完成特征选择

```python
from sklearn.feature_selection import RFECV
#estimator:要操作的模型
#step:每次删除的变量数
#cv:使用的交叉验证折数
#n_jobs:并发的数量
#scoring:评估的方法
rfecv=RFECV(estimator=lr,step=1,cv=5,n_jobs=-1,scoring='r2')
rfecv.fit(x_train_scale,y_train)
print('选择后剩余的特征数量',rfecv.n_features_)
print('选择后使用缩减特征训练后的模型',rfecv.estimator_)
print('每个特征的等级，数值越小特征越重要',rfecv.ranking_)
print('特征是否被选择',rfecv.support_)
print('模型交叉验证的评分',rfecv.grid_scores_)
```

```python
#绘制在特征选择过程中使用交叉验证得到的R^2值
plt.plot(range(1,len(rfecv.grid_scores_)+1),rfecv.grid_scores_,marker='o')
plt.xlabel('特征数量')
plt.ylabel('交叉验证R^2')
plt.show()
```

```python
print('剔除的变量',X.columns.values[~rfecv.support_])
x_train_eli=rfecv.transform(x_train_scale)
x_test_eli=rfecv.transform(x_test_scale)
print('训练数据得分',rfecv.estimator_.score(x_train_eli,y_train))
print('测试数据得分',rfecv.estimator_.score(x_test_eli,y_test))
```

```python
"""
在python中有pandas的get_dummies和sklearn（from sklearn.preprocessing import OneHotEncoder）
（遗憾的是sklearn中的OneHotEncoder无法直接对字符串型的类别变量编码，可以对数值型编码；而get_dummies对列是字符型还是数字型都可以进行二值编码）
get_dummies 不像 sklearn 的 transformer一样，有 transform方法，
所以一旦测试集中出现了训练集未曾出现过的特征取值，简单地对测试集、训练集都用 get_dummies 方法将导致数据错误。
"""
```

```python
#one_hot 类别变量中n个不同类别转换为n个变量 
#dummy variable 在某一设定的参考准则下，对n个不同的类别，转换为n-1个变量
#from sklearn.preprocessing import LabelEncoder  #类别型
#from sklearn.preprocessing import OneHotEncoder #数值型
```

```python
from sklearn.preprocessing import KBinsDiscretizer
#KBinsDiscretizer K个分箱的离散器，用于将数值（连续型数值）进行离散化
#n_bins:分箱（区间）的个数
#encode:离散化编码方式
    #onehot:独热编码：返回稀疏矩阵，只有1，没有0
    #onehot-dense:独热编码：返回稠密矩阵，有0，1
    #ordinal:使用序数编码
#strategy:分箱的方式，分为uniform,quantile,keams
    #uniform:每个区间的长度范围大致相同
    #quantile:每个区间包含的元素个数大致相同
    #kmeans：
k=KBinsDiscretizer(n_bins=[4,5,10,6],encode='onehot-dense',strategy='uniform')
#离散化特征
discretize=['Longitude','Temperature','Precipitation','Latitude']

x_train_eli=pd.DataFrame(data=x_train_eli,columns=X.columns[rfecv.support_])
x_test_eli=pd.DataFrame(data=x_test_eli,columns=X.columns[rfecv.support_])

r=k.fit_transform(x_train_eli[discretize])
r=pd.DataFrame(r,index=x_train_eli.index)
```

```python
r
```

```python
x_train_eli
```

```python
#获取除离散化特征之外的其他特征
x_train_dis=x_train_eli.drop(discretize,axis=1)
#将离散化的特征与其他特征进行组合
x_train_dis=pd.concat([x_train_dis,r],axis=1)
x_train_dis
```

```python
#对测试集做同样的离散化处理
r=pd.DataFrame(k.transform(x_test_eli[discretize]),index=x_test_eli.index)
x_test_dis=x_test_eli.drop(discretize,axis=1)
x_test_dis=pd.concat([x_test_dis,r],axis=1)
```

```python
lr.fit(x_train_dis,y_train)
print('训练数据得分',lr.score(x_train_dis,y_train))
print('测试数据得分',lr.score(x_test_dis,y_test))
```

- 可见，离散化后模型效果有了一定提升

#### 残差图分析
- 残差，即模型预测值与真实值之间的差异
- 如果能从残差图中找出某种规律则表示还有能解释因变量的信息没有提取
- 残差图的作用：
    - 检测模型是否具有异方差性
    - 检测异常值（离群点）
    - 检测残差的分布
    - 检测自变量与因变量是否为线性关系

##### 异方差性处理

```python
fig,ax=plt.subplots(1,2)
fig.set_size_inches(15,5)
data=[x_train,x_train_dis]
title=['原始数据','处理后数据']
for d,a,t in zip(data,ax,title):
    model=LinearRegression()
    model.fit(d,y_train)
    y_hat_train=model.predict(d)
    residual=y_hat_train-y_train
    a.set_xlabel('预测值')
    a.set_ylabel('残差')
    a.set_title(t)
#     a.scatter(x=y_hat_train,y=residual)
    sns.scatterplot(x=y_hat_train,y=residual,ax=a)
    a.axhline(y=0,color='red')
```

- 从图中发现，随着预测值的增大，模型的误差也在增大，对于此种情况，可以使用对目标y值取对数的方式进行处理

```python
model=LinearRegression()
y_train_log=np.log(y_train)
y_test_log=np.log(y_test)
model.fit(x_train_dis,y_train_log)

y_hat_train=model.predict(x_train_dis)
residual=y_hat_train-y_train_log.values
sns.scatterplot(x=y_hat_train,y=residual)
plt.xlabel('预测值')
plt.ylabel('残差')
plt.axhline(y=0,color='red')
plt.show()
```

```python
print('训练集得分',model.score(x_train_dis,y_train_log))
print('测试集得分',model.score(x_test_dis,y_test_log))
```

- 具体情况具体判断，这里模型的效果并没有得到提升

##### 离群点处理

```python
#依然采用x_train_dis和y_train训练的模型lr
y_hat_train=lr.predict(x_train_dis)
residual=y_hat_train-y_train.values

r=(residual-residual.mean())/residual.std()#几倍标准差
```

```python
where=abs(r)<=2
sns.scatterplot(x=y_hat_train[where],y=residual[where],color='b',label='正常值')
sns.scatterplot(x=y_hat_train[~where],y=residual[~where],color='orange',label='异常值')
plt.axhline(y=0,color='red')
plt.xlabel('预测值')
plt.ylabel('残差')
plt.legend()
plt.show()
```

```python
#剔除离群点
x_train_filter=x_train_dis[where]
y_train_filter=y_train[where]
lr.fit(x_train_filter,y_train_filter)
print('训练集得分',lr.score(x_train_filter,y_train_filter))
print('测试集得分',lr.score(x_test_dis,y_test))
```
